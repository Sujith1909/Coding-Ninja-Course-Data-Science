```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
```

```python
x1 = np.array([1,2,3,4,5,6])
x2 = np.array([7.5,11,16,18,20,26])
```

```python
plt.scatter(x2,x1)
plt.show()
```

```python
x = np.c_[x1,x2]
```

```python
x
```

```python
pca = PCA(n_components = 2)
```

```python
xTransformed = pca.fit_transform(x)
```

```python
xTransformed
```

```python
pca.components_
#these are the two unit vectors in these two directions.
#these are the new two axes in which data is stored 
```

```python
pca.inverse_transform(xTransformed)#to transfrom the data back to the original
```

```python
np.random.seed(2332321)
```

```python
mean_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1 = np.random.multivariate_normal(mean_vec1,cov_mat1,100)
```

```python
mean_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2 = np.random.multivariate_normal(mean_vec2,cov_mat2,100)
```

```python
from mpl_toolkits.mplot3d import Axes3D, proj3d
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111,projection ='3d')
ax.plot(class1[:,0],class1[:,1],class1[:,2],'o')
ax.plot(class2[:,0],class2[:,1],class2[:,2],'^')
```

### Now we change this 3d data into 2d data 

```python
pca = PCA(n_components = 2)
```

```python
combined = np.concatenate((class1,class2))
```

```python
transformed_data = pca.fit_transform(combined)
```

```python
transformed_data
```

```python
pca.components_
# pca saying this this 2 dimensions are important than the other dimension
```

```python
plt.plot(transformed_data[:100,0],transformed_data[:100,1],'o')
plt.plot(transformed_data[100:200,0],transformed_data[100:200,1],'^')
plt.show()
```

```python
x_approx = pca.inverse_transform(transformed_data)
x_approx
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111,projection ='3d')
ax.plot(x_approx[:,0],x_approx[:,1],x_approx[:,2],'o')
#ax.plot(x_approx[:,0],x_approx[:,1],x_approx[:,2],'^')
```
