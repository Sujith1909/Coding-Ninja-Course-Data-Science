```python
def cost(m, x, y):
    cost=0
    for i in range(len(x)):
        cost+=(1/len(x))*((y[i]-sum(m*x[i]))**2)
    print(cost)
```

```python
def step_gradient(X, Y, m, alpha):
    m_slope=np.zeros(len(X[0]))
    for i in range(len(X)):
        x=X[i]
        y=Y[i]
        for j in range(len(x)):
            m_slope[j]+=(-2/len(X))*(y-sum(m*x))*x[j]
    new_m=m-(alpha*m_slope)
    return new_m
```

```python
def gd(x, y, alpha, iterations):
    m=np.zeros(len(x[0]))
    for i in range(iterations):
        m=step_gradient(x, y, m,alpha)
    return m

```

```python
def gradient_descent(x, y):
    iterations=200
    alpha=0.1
    x=np.append(x, np.ones(len(x)).reshape(-1, 1), axis=1)# for the intercept coloumn
    m=gd(x, y, alpha, iterations)
    return m
```
