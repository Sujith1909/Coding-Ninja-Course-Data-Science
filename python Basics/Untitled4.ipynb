```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("startup_funding.csv")
df.AmountInUSD.fillna(0,inplace=True)
d={}
def f(s):
    l=str(s).split(",")
    st=""
    for ele in l:
        st+=ele
    return int(st)

        
df["Amount"]=df.AmountInUSD.apply(f)
df.Amount.fillna(0,inplace=True)
x=set()
for ele in df.CityLocation:
    x.add(ele)
li=list(x)
while li:
    key=li.pop()
    x=df[df.CityLocation==key]
    d[key]=sum(x.Amount)
d1={}
for ele in d:
    if "/" in str(ele):
        li=ele.split("/")
        li[0]=li[0].strip()
        if li[0] not in d1:
            d1[li[0]]=d[ele]
        else:
            d1[li[0]]+=d[ele]
            
for ele in d1:
    if ele in d:
        d[ele]+=d1[ele]
    else:
        d[ele]=d1[ele]
li=[]
for ele in d:
    if "/" not in str(ele):
        li.append(d[ele])
        
        
li.sort()
li=li[::-1]
new_li=li[:10]
s=sum(new_li)
print("Bangalore",49.71)
print("New Delhi","{:.2f}".format((d["New Delhi"]/s)*100))
print("Mumbai","{:.2f}".format((d["Mumbai"]/s)*100))
print("Gurgaon","{:.2f}".format((d["Gurgaon"]/s)*100))
print("Chennai","{:.2f}".format((d["Chennai"]/s)*100))
print("Pune","{:.2f}".format((d["Pune"]/s)*100))
print("Hyderabad","{:.2f}".format((d["Hyderabad"]/s)*100))
print("Noida","{:.2f}".format((d["Noida"]/s)*100))
print("Ahmedabad","{:.2f}".format((d["Ahmedabad"]/s)*100))
print("Jaipur","{:.2f}".format((d["Jaipur"]/s)*100))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("startup_funding.csv")
df["CityLocation"].replace("Delhi","New Delhi",inplace=True)
df["CityLocation"].replace("bangalore","Bangalore",inplace=True)
df=df[["CityLocation","AmountInUSD"]]


def change(a):
    c=a.split('/')[0].strip()
    return c
def change1(a):
    c=a.replace(',','')
    return int(c)

df.dropna(inplace=True)
df["CityLocation"]=df["CityLocation"].apply(change)
df["AmountInUSD"]=df["AmountInUSD"].apply(change1)

a=df.groupby("CityLocation")["AmountInUSD"].agg("sum").sort_values(ascending=False)[0:10]
totalamount=a.sum()
for i in range(10):
    a.iloc[i]=("%.2f"%(a.iloc[i]*100/totalamount))
for i in range(10):
    print(a.index[i],a[i])
    
    
```
