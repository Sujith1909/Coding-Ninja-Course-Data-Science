# Question1

```python
import pandas as pd
import operator
import matplotlib.pyplot as plt
df = pd.read_csv('startup_funding.csv')

df["CityLocation"]= df["CityLocation"].str.split("/", n = 1, expand = True)#n=1 for only one time eg banglore/kolkata/delhi ==banglore,/kolkata/delhi
df["CityLocation"]= df["CityLocation"].str.strip()# to remove white spaces after variable 
df["CityLocation"]= df["CityLocation"].replace("bangalore","Bangalore")
df["CityLocation"]= df["CityLocation"].replace("Delhi","New Delhi")
df["CityLocation"]= df["CityLocation"].replace("New Delhi","NCR")
df["CityLocation"]= df["CityLocation"].replace("Noida","NCR")
df["CityLocation"]= df["CityLocation"].replace("Gurgaon","NCR")
## amount is change to float type and , is removed 
df['AmountInUSD']=df['AmountInUSD'].str.replace(',','')
df[['AmountInUSD']] = df[['AmountInUSD']].apply(pd.to_numeric) 
## dataframe with required city
citydf=df[(df['CityLocation']=='Bangalore')|(df['CityLocation']=='Mumbai')|(df['CityLocation']=='NCR')]

## dictionary is made to store the frequency
dict={}
for i in citydf.CityLocation:
    dict[i]=dict.get(i,0)+1
sorted_tuples = sorted(dict.items(), key=operator.itemgetter(1),reverse=True)
dict = {k: i for k, i in sorted_tuples}
print('The City he should select is:',list(dict.keys())[0],'|',list(dict.values())[0])
number=[]
city=[]
for key in dict:
    number.append(dict[key])
    city.append(key)
plt.bar(city,number,edgecolor='Black')
plt.xlabel('City')
plt.ylabel('Number of Funding Receive')
plt.xticks(rotation=45)
plt.show()
print("\033[1m"+"Conclusion:"+"\033[0m"+"\nFrom above bar chart it seems that maximum number of fundings has been raised for NCR which is 709.")
print("I should suggest my friend to select location NCR as first choice followed by Banglore and Mumbai")
```

# Question 2

```python
import matplotlib.pyplot as plt
import pandas as pd
import csv
df=pd.read_csv("startup_funding.csv")
df=df.copy()
invester=df['InvestorsName']
dict={}
for i in invester :
    p = str(i)
    q = p.split(",")
    for j in q:
        j = j.strip()
        if j in dict:
            dict[j]+=1
        else:
            dict[j] = 1
sort_orders = sorted(dict.items(), key=lambda x: x[1], reverse=True)
print("\033[1m"+"Conclusion:"+"\033[0m"+"\nFrom above list we can advise friend to contat these top 5 investors for funding")
print("\033[1m"+"\nTop 5 Investor:"+"\033[0m")
print()
print("Investor  \t      |   No of times invested")
print()
investor=[]
count=[]
for i in sort_orders[0:5]:
    if i[0]!='':
        print('{:<22s}|{:>11d}'.format(i[0],i[1]))
        investor.append(i[0])
        count.append(i[1])


color=['Blue','Orange','Green','Red','Purple']
plt.pie(count,labels=investor,autopct="%.2f")
plt.show()
plt.bar(investor,count,edgecolor='Black')
plt.xlabel('INVESTORS')
plt.ylabel('NO Of TIMES INVESTED')
plt.xticks(rotation=30)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import csv
df=pd.read_csv("startup_funding.csv")
df=df.copy()
df['StartupName'].replace('Flipkart.com','Flipkart',inplace=True)
df['StartupName'].replace('Ola Cabs','Ola',inplace=True)
df['StartupName'].replace('Olacabs','Ola',inplace=True)
df['StartupName'].replace('Oyo Rooms','Oyo',inplace=True)
df['StartupName'].replace('Oyorooms','Oyo',inplace=True)
df['StartupName'].replace('OyoRooms','Oyo',inplace=True)
df['StartupName'].replace('OYO Rooms','Oyo',inplace=True)
df['StartupName'].replace('Paytm Marketplace','Paytm',inplace=True)
df['InvestorsName'].replace('Undisclosed investors','Undisclosed Investors',inplace=True)
df['InvestorsName'].replace('undisclosed investors','Undisclosed Investors',inplace=True)
df['InvestorsName'].replace('Undisclosed Investor','Undisclosed Investors',inplace=True)
df['InvestorsName'].replace('Undisclosed','Undisclosed Investors',inplace=True)
df['InvestorsName'].replace('Undisclosed investor','Undisclosed Investors',inplace=True)
df['InvestorsName'].replace('Undisclosed Investors','',inplace=True)
df['StartupName'].fillna('Unknown',inplace=True)
df['InvestorsName'].fillna('Unknown',inplace=True)
investor =df['InvestorsName']
startup=df['StartupName']
dict={}
k=0
for i in investor:
    p= str(i)
    j=p.split(',')
    for ele in j:
        ele=ele.strip()
        if ele in dict:
            dict[ele]+=[startup[k]]
        else:
            dict[ele]=[startup[k]]
    k=k+1
for ele in dict:
    dict[ele]=(len(set(dict[ele]))) 
sort_orders = sorted(dict.items(), key=lambda x: x[1], reverse=True)
print("Investor  \t      |   No of times invested")
print()
investor=[]
count=[]
for i in sort_orders[0:6]:
    if i[0]!='':
        print('{:<22s}|{:>11d}'.format(i[0],i[1]))
        investor.append(i[0])
        count.append(i[1])


color=['Blue','Orange','Green','Red','Purple']
plt.pie(count,labels=investor,autopct="%.2f")
plt.show()
plt.bar(investor,count,edgecolor='Black')
plt.xlabel('INVESTORS')
plt.ylabel('NO Of TIMES INVESTED')
plt.xticks(rotation=30)
plt.show()
print("\033[1m"+"Conclusion:"+"\033[0m"+"\nWe should suggest friend to get in touch with above top 5 investors on the basis of ivestment in unique startups.\n")
```
