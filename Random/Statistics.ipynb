```python
import pandas as pd 
```

```python
df = pd.read_csv('0000000000002370.csv')
```

```python
df.head()
```

```python
df = df[['CityLocation','AmountInUSD']]
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
for i in df['AmountInUSD']:
    print(type(i))
```

```python
for i in range(len(df['AmountInUSD'])):
    df['AmountInUSD'][i] = int(df['AmountInUSD'][i].replace(',',''))
```

```python
avg = df['AmountInUSD'].mean()
print(avg)
```

```python
df["CityLocation"]= df["CityLocation"].replace("bangalore","Bangalore")
df["CityLocation"]= df["CityLocation"].replace("Delhi","New Delhi")
```

```python
avg = df['AmountInUSD'].mean()
print(avg)
avg = 12284548.63
```

```python
citydf=df[(df['CityLocation']=='Bangalore')|(df['CityLocation']=='Delhi')]
```

```python
avg = citydf['AmountInUSD'].mean()
print(avg)
```

```python
avg_sample =citydf['AmountInUSD'].sample(50,random_state = 1).mean()
print(avg_sample)
```

```python
avg_sample_replacment =citydf['AmountInUSD'].sample(75,replace = True).mean()
print(avg_sample_replacment)
```

```python
population_sampling_error = avg - avg_sample
print(population_sampling_error)
```

```python
population_sampling_error_replacment = avg - avg_sample_replacment
print(population_sampling_error_replacment)
```
